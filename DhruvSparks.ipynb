##TASK:-Predict the Precentage of an student based on the number of study hours, perform Prediction using Supervised Machine Learning.

##Step 1:- Importaing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
print("libraries imported")
```

##Step 2:- Loading the dataset and describing it. 

```python
url="http://bit.ly/w-data"
df=pd.read_csv(url)
print("data loaded")
print(df)
```

```python
df.describe()
```

##Step 3:-Visualizing the data using Statter plot

```python
plt.scatter(df['Hours'],df['Scores'],color="red")
plt.xlabel("Hours studied")
plt.ylabel("Percentage scored")
plt.title("Relation between hours studied vs percentage scored")
plt.show()
```

##Step 4:- Separate Dependent and independent variables 

```python
x=df['Hours'].values.reshape(-1,1)
y=df['Scores'].values.reshape(-1,1)
```

##Step 5:-Spliting the data into traning and testing dataset

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

##Step 6:- Training the model

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

##Step 7:-Obtaining the intercept and slope of regression line

```python
print("slope", regressor.coef_)
print("intercept",regressor.intercept_)
```

##Step 8:-Plotting the regression line in Scatter plot

```python
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,x_train*regressor.coef_+regressor.intercept_,color="blue")
plt.show()
```

##Step 9:-Predicting the Data and visulizing the relationship between actual and predicted data

```python
y_pred=regressor.predict(x_test)
df1=pd.DataFrame({"Actual":y_test.flatten(),"Predicted":y_pred.flatten()})
df1.plot(kind="bar")
plt.show()
```

##Step 10:-Efficiency of model

```python
print("Mean absolute Error ", metrics.mean_absolute_error(y_test,y_pred))
```

#Internal Task:-To predict the percentage of student who studied for 9.25 hours/day 

```python
predicted_val=regressor.predict([[9.25]])
print("Score for this student will be:",predicted_val.flatten())
```
